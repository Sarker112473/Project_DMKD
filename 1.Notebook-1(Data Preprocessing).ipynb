## **Data Transformation**



### **Min-Max Normalization**



\begin{equation}
X^{*} = \frac{X-min(x)}{range(X)} = \frac{X-min(X)}{max(X)-min(X)}
\end{equation}



**Min-Max Normalization of CO2:**

Let's check the current condition of the field before any normalization:

```python
df['CO2'].describe()
```

Let's apply the equation that we have formulated for finding the normalized value:

```python
normalized_df = df

normalized_df['CO2']=(df['CO2']-df['CO2'].min())/(df['CO2'].max()-df['CO2'].min())
```

Let's check the normalized value for the column acceleration:


```python
normalized_df['CO2'].describe()
```

### **Z-Score Standardization**



\begin{equation}
    X^{*} = \frac{X-mean(x)}{SD(X)}
\end{equation}



**Z-Score Standardization of CO2:**

```python
standardized_df = df

standardized_df['CO2']=(df['CO2']-df['CO2'].mean())/(df['CO2'].std())
```

```python
standardized_df['CO2'].describe()
```

```python
standardized_df = df

standardized_df['Light']=(df['Light']-df['Light'].mean())/(df['Light'].std())
```

### **Outlier detection:**


**CO2:**

```python
import numpy as np
import matplotlib.pylab as plt


fig = plt.figure(1, figsize=(6, 7))
ax = fig.add_subplot(111)

ax.set_title('Outliers can be seen both above and below the IQR')

ax.boxplot(df['CO2'], vert=True, manage_ticks=True)
ax.set_ylabel('Inter Quantile Range Values')
# ax.set_yticklabels(['Min','Q1','Median', 'Q3','Max'])
ax.set_xticklabels(['CO2'])

quantiles = np.quantile(df['CO2'], np.array([0.00, 0.25, 0.50, 0.75, 1.00]))
ax.hlines(quantiles, [0] * quantiles.size, [1] * quantiles.size,
          color='b', ls=':', lw=0.5, zorder=0)
ax.set_xlim(0.5, 1.5)
ax.set_yticks(quantiles)
plt.show()
```
